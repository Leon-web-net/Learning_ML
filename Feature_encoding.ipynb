# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

def LE_func(df, col):
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    return le
```

```python
le_dict= {}
```

```python

cols = ["sex", "embarked"]
for col in cols:
    if isinstance(df_X[col].dtype, pd.CategoricalDtype) or pd.api.types.is_object_dtype(df_X[col]):
      le_dict[col] = LE_func(df_X, col)
    else:
      print(f"{col} is of {df_X[col].dtype} and has not been labelEncoded")

df_X.head()
```

```python
for col, le in le_dict.items():
    mapping = {label: idx for idx, label in enumerate(le.classes_)}
    print(f"{col}: {mapping}")
```

# Ordinal Encoding

```python
from sklearn.datasets import fetch_openml

data = fetch_openml(name="credit-g", version=1, as_frame=True)
```

```python
df_y = pd.DataFrame(data.target)
df_X = pd.DataFrame(data.data)

df_X.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

def OE_func(df, col,cats):
    oe = OrdinalEncoder(categories=[cats])
    df[col] = oe.fit_transform(df[[col]]).astype(int)
    return oe
```

```python
oe_dict = {}
```

```python
cols = ["checking_status", "credit_history"]
[df_X[col].unique() for col in cols]
```

```python
category = {"checking_status": ['<0','no checking','0<=X<200','>=200'],
            "credit_history":['critical/other existing credit','delayed previously', 'existing paid', 'no credits/all paid', 'all paid']}
for col in cols:
    if isinstance(df_X[col].dtype, pd.CategoricalDtype) or pd.api.types.is_object_dtype(df_X[col]):
       oe_dict[col] = OE_func(df_X, col,category[col])
    else:
      print(f"{col} is of {df_X[col].dtype} and has not been labelEncoded")

df_X.head()
```

```python
for col, oe in oe_dict.items():
    mapping = {label: idx for idx, label in enumerate(oe.categories_[0])}
    print(f"{col}: {mapping}")
```
